```python
import pandas as pd
import numpy as np

# Load CSV files
df1 = pd.read_csv(r"C:\Users\black\Documents\centraleSupelec\3A\cours\DL&NLP\Projet_final\submission_ATP_PB.csv")
df2 = pd.read_csv(r"C:\Users\black\Desktop\DATASET\subset_test_ATPvalues.csv", sep=";")
df2.head()
```

```python

# Merge on common key
df = pd.merge(df1, df2, on="pid", how="inner")

# Optional: inspect columns after merge
print(df.columns)
```

```python

# Avoid division by zero
df = df[df["atp_value_raw_x"] != 0]

# Calculate Absolute Percentage Error (APE)
df["ape"] = np.abs((df["atp_value_raw_x"] - df["atp_value_raw_y"]) / df["atp_value_raw_x"])

# Calculate MAPE
mape = df["ape"].mean() * 100

print(f"MAPE: {mape:.2f}%")
```
